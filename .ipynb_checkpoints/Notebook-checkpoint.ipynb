```python
import statsmodels.api as sm
import statsmodels.tsa.api as tsa
import statsmodels.formula.api as smf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from arch import arch_model
import statsmodels
```

```python
plt.rcParams["figure.figsize"] = (14, 8)
```

```python
df = pd.read_csv('spy.txt')
spy = pd.Series(list(df['return']), index=pd.to_datetime(df['date']))#.asfreq('D', 'pad')

df = pd.read_csv('spy1m.txt')
spy1m = pd.Series(list(df['return']), index=pd.to_datetime(df['date']), name='data').asfreq('M', 'pad')

spy1w = np.expm1(np.log1p(spy).resample('W').sum())
```

```python
#AR
pd.Series(tsa.acf(spy1m, nlags=15)).plot(label='Monthly S&P 500').legend()
pd.Series(tsa.acf(spy, nlags=15)).plot(label='Daily S&P 500').legend()
pd.Series(tsa.acf(spy1w, nlags=15)).plot(label='Weekly S&P 500').legend()
```

```python
#MA
pd.Series(tsa.pacf(spy1m, nlags=15)).plot(label='Monthly S&P 500').legend()
pd.Series(tsa.pacf(spy, nlags=15)).plot(label='Daily S&P 500').legend()
pd.Series(tsa.acf(spy1w, nlags=15)).plot(label='Weekly S&P 500').legend()
```

```python
nobs = len(spy1m)
ntrain = int(len(spy1m)*0.5)
train = spy1m[:ntrain]
mod = tsa.SARIMAX(train, order=(1, 0, 0))
res = mod.fit()
res1 = res
forecasts = []
forecasts.append(res.forecast(steps=1))

for i in range(ntrain, nobs):
    res = res.extend(spy1m.iloc[i:i+1])
    fc = res.forecast(steps=1)
    forecasts.append(fc)

fcs = pd.concat(forecasts)
print(fcs.corr(spy1m))
fcs.plot()
spy1m.plot()
```

```python
def run(xs, order):
    df = xs.to_frame() 
    win = 96
    import warnings
    warnings.filterwarnings("ignore")
    mod = tsa.SARIMAX(df[:win], order=(1, 0, 1))
    res = mod.fit()
    fcasts = []
    fcasts.append(res.forecast())
    
    for i in range(win, len(df)):
        res = res.append(df.iloc[i:i+1], refit=True)
        fcasts.append(res.forecast())
    fcs = pd.concat(fcasts)
    fcs.plot(label='Forecasted S&P 500 Return').legend()
    xs[xs.index >= fcs.index[0]].plot(label='S&P 500 Return').legend()
    y_pred = fcs
    y_true = xs[xs.index >= fcs.index[0]] 
    ys = (y_true * y_pred > 0).value_counts().values
    print('Corr', xs.corr(fcs))
    print('Accuracy', ys[0]/sum(ys))
    
    return fcs
```

```python
run(spy1m, (1, 0, 1))
```

```python
am = arch_model(spy, rescale=True, dist='t')
index = spy.index
start_loc = 0
end_loc = 63
forecasts = {}
for i in range(len(spy) - end_loc):
    res = am.fit(first_obs=0, last_obs=i + end_loc, disp='off')
    temp = np.sqrt(res.forecast(horizon=1).variance)
    fcast = temp.iloc[i + end_loc - 1]
    forecasts[fcast.name] = fcast
```

```python
arch_vols = np.sqrt(252)*(pd.DataFrame(forecasts).T / am.scale).shift(1).iloc[1:]['h.1']
spy_vols = np.sqrt(252)*spy.rolling(21).std().iloc[end_loc:]
```

```python
arch_vols.plot(label='Annualized Conditional Volatility').legend()
spy_vols.plot(label='Annualized 21d Conditional Volatility').legend()
```

```python
print("MAPE:", np.mean(np.abs((spy_vols - arch_vols) / spy_vols)))
print("Corr:", arch_vols.corr(spy_vols))
```
